```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../dataset/NFL_09-10.csv")
np.random.seed(0) 
```

```python
df.shape
```

```python
df.head()
```

```python
missing_values_count = df.isnull().sum()
missing_values_count
```

```python
missing_values_count[0:13]
```

```python
# how many total missing values do we have?
total_cells = np.product(df.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

```python
###can try to guess what it might have the values been based on the other values in that column and row - called imputation
```

```python
#Drop missing values#
# remove all the rows that contain a missing value
##df.dropna()
#This is not applicable always
```

```python
# remove all columns with at least one missing value
columns_with_na_dropped = df.dropna(axis=1)
columns_with_na_dropped.head()
#this method also not applicable always becz we loss large data.
```

```python
columns_with_na_dropped.shape
```

```python
#Let's look the amount of data we lost.
print("Columns in original dataset: %d \n" % df.shape[1])
print("Columns with na's dropped: %d" % columns_with_na_dropped.shape[1])
```

```python
#Filling in missing values automatically
# get a small subset of the dataset
subset_nfl_data = df.loc[:, 'EPA':'Season'].head()
subset_nfl_data
```

```python
#We can use the Panda's fillna() function to fill in missing values in a dataframe for us
# replace all NA's with 0
subset_nfl_data.fillna(0)
```

```python
#in the above method make us fill zero values more
#so let us do some logical methods like bfill or ffill
```

```python
# replace all NA's the value that comes directly after it in the same column, 
# then replace all the remaining na's with 0
subset_nfl_data.fillna(method='bfill', axis=0).fillna(0)
```
